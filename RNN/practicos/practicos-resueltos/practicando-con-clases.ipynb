```python
class NumeroComplejo:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def suma(self,c):
        return NumeroComplejo(self.a+c.a,self.b+c.b)
    def real(self):
        return self.a
    def imaginaria(self):
        return self.b
    def __add__(self,c):
        return NumeroComplejo(self.a+c.a,self.b+c.b)
    def __repr__(self):
        return self.__str__()
    def __str__(self):
        return str(self.a) + ' + i*' + str(self.b)  
```
